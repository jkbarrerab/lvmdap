```python
import sys
import numpy as np
from os.path import join
import matplotlib.pyplot as plt
from copy import deepcopy as copy

from pyFIT3D.common.stats import median_filter
from pyFIT3D.modelling.stellar import StPopSynt
from pyFIT3D.common.auto_ssp_tools import ConfigAutoSSP
from pyFIT3D.common.io import remove_isfile, read_spectra
from pyFIT3D.common.constants import __c__, __sigma_to_FWHM__
from pyFIT3D.common.io import create_ConfigAutoSSP_from_lists
from pyFIT3D.common.io import create_emission_lines_file_from_list
from pyFIT3D.common.gas_tools import detect_create_ConfigEmissionModel
from pyFIT3D.common.io import create_emission_lines_mask_file_from_list
```

```python
output_path = '/home/lacerda/tests/detect_lines/'
name = 'LMC_043-1'
spec_filename = join(output_path, f'{name}.txt')
wave, flux, eflux = read_spectra(spec_filename, variance_column=False)
```

```python
plt.gcf().set_size_inches(15,5)
ax = plt.gca()
ax.set_ylabel(r'Flux', fontsize=18)
ax.set_xlabel(r'wavelength [\AA]', fontsize=18)
ax.plot(wave, flux, color='k', label='obs')
ax.plot(wave, eflux, color='r', label='error')

new_eflux = 0.3*median_filter(int(10*__sigma_to_FWHM__), copy(flux))
ax.plot(wave, new_eflux, color='y', label='new error')

ax.legend(fontsize=20)
```

```python
redshift = 0.00012
sigma_guess = 2.55
chunks = 3
     chunks = [[4700, 5500], [5500, 7000], [7000, 9349]]
#chunks = np.asarray(chunks)
peak_find_nsearch = 1
peak_find_threshold = 0.5
peak_find_dmin = 1
plot = 1
crossmatch_list_filename = None
crossmatch_list_filename = '/home/lacerda/tests/detect_lines/emission_lines_long_list.MaNGA'
#crossmatch_list_filename = '/home/lacerda/tests/detect_lines/emission_lines.txt'
crossmatch_min_dist_AA = 4
sort_by_flux = True
verbose = 2
```

```python
r = detect_create_ConfigEmissionModel(
    
    # spectrum
    wave, flux, 
    
    redshift=redshift,
    sigma_guess=sigma_guess,
    
    # Chunks to be analyzed (i.e. the chunks of the generated ConfigEmissionModel files)
    chunks=chunks,
    
    # Add a polynomial model to the generated configs
    polynomial_order=2,
    polynomial_coeff_guess=[0.001, 0.001],
    polynomial_coeff_boundaries=[[-1e13, 1e13], [-1e13, 1e13]],

#     polynomial_order=1,
#     polynomial_coeff_guess=[0.001],
#     polynomial_coeff_boundaries=[-1e13, 1e13],

    ########################################################
    ### DEFAULT VALUES, NOT NEED TO BE USED FOR THOSE VALUES
    flux_boundaries_fact=[0.001, 1000],
    sigma_boundaries_fact=[0.1, 2],
    v0_boundaries_add=[-50, 50],
    ########################################################
    
    # Peak find options
    peak_find_nsearch=peak_find_nsearch,
    peak_find_threshold=peak_find_threshold,
    peak_find_dmin=peak_find_dmin,
    
    
    # Crossmatch options
    crossmatch_list_filename=crossmatch_list_filename,
    crossmatch_min_dist_AA=crossmatch_min_dist_AA,

    # Nice plots
    plot=plot,
    
    output_path=output_path,
    sort_by_flux=sort_by_flux,
    
    verbose=verbose,
)
config_filenames, wl_chunks, wave_peaks_tot, wave_peaks_tot_corr = r
```

```python
create_emission_lines_mask_file_from_list(wave_peaks_tot, eline_half_range=8*sigma_guess, output_path=output_path)
create_emission_lines_file_from_list(wave_peaks_tot, output_path=output_path)
create_ConfigAutoSSP_from_lists(wl_chunks, config_filenames, output_path=output_path)
```

```python
#####################################################################
#
# --= pyFIT3D Auto SSP analysis =--
#

sigma_inst = 2.3
mask_list = None
w_min_max = [4750, 9300]
nl_w_min_max = [4750, 5700]
ssp_nl_fit_file = '/home/lacerda/LOCAL/data/models/gsd01_3.fits'
ssp_file = '/home/lacerda/LOCAL/data/models/gsd01_156.fits'
elines_mask_file = join(output_path, 'autodetect.emission_lines.txt')
redshift_set = [0.0002, 0.0001, -0.001, 0.001]
sigma_set = [30, 1, 0, 80]
AV_set = [0.1, 0.1, 0, 1.6]
config_file = join(output_path, 'autodetect.auto_ssp_several.config')

# final output file:
out_file = f'auto_ssp.{name}.int.out'
out_file_elines = 'elines_' + out_file
out_file_coeffs = 'coeffs_' + out_file
out_file_fit = 'output.' + out_file + '.fits'
remove_isfile(out_file)
remove_isfile(out_file_elines)
remove_isfile(out_file_coeffs)
remove_isfile(out_file_fit)

cfAutoSSP = ConfigAutoSSP(config_file, redshift_set=redshift_set, sigma_set=sigma_set, AV_set=AV_set)
SPS = StPopSynt(config=cfAutoSSP,
                wavelength=wave, flux=flux, eflux=new_eflux,
                mask_list=mask_list, 
                elines_mask_file=elines_mask_file,
                sigma_inst=sigma_inst, 
                ssp_file=ssp_file,
                ssp_nl_fit_file=ssp_nl_fit_file, 
                out_file=out_file,
                w_min=w_min_max[0], w_max=w_min_max[1], 
                nl_w_min=nl_w_min_max[0], nl_w_max=nl_w_min_max[1],
                R_V=3.1, extlaw='CCM')
```

```python
msg_cut = f' - cut value: {cfAutoSSP.CUT_MEDIAN_FLUX:6.4f}'
if cfAutoSSP.CUT_MEDIAN_FLUX == 0:
    msg_cut = ' - Warning: no cut (CUT_MEDIAN_FLUX = 0)'
print(f'-> median raw flux = {SPS.median_flux:6.4f}{msg_cut}')
```

```python
SPS.non_linear_fit()
```

```python
SPS.calc_SN_norm_window()
```

```python
# SPS.gas_fit(ratio=True)
ssp_models = SPS.models
ssp_coeffs = SPS.get_last_coeffs_ssp()
model_min = SPS.get_best_model_from_coeffs(ssp=ssp_models, coeffs=ssp_coeffs)
print(SPS.SN_norm_window)
if SPS.SN_norm_window > 10:
    SPS._subtract_continuum(model_min)
```

```python
plt.gcf().set_size_inches(15, 5)
plt.plot(SPS.spectra['orig_wave'], SPS.spectra['orig_flux'], label='obs', color='k', lw=3)
plt.plot(SPS.spectra['orig_wave'], model_min, label='model', color='y', lw=1)
plt.plot(SPS.spectra['orig_wave'], SPS.spectra['orig_flux'] - model_min, label='residual', color='r')
plt.legend()
```

```python
SPS._EL_fit(model_min=model_min)
```

```python
s = SPS.spectra

s['raw_flux_no_gas'] = s['orig_flux'] - s['raw_model_elines']

y_ratio = SPS._rescale(model_min)

ratio = np.divide(s['orig_flux'], y_ratio, where=y_ratio!=0)
s['orig_flux_ratio'] = np.where(y_ratio > 0, ratio, s['orig_flux'])
ratio = np.divide(s['raw_flux_no_gas'], y_ratio, where=y_ratio!=0)
s['raw_flux_no_gas'] = np.where(y_ratio > 0, ratio, s['raw_flux_no_gas'])
```

```python
plt.gcf().set_size_inches(15, 5)
mask = s['raw_wave'] < 9500
plt.plot(s['raw_wave'][mask], s['raw_model_elines'][mask])
```

```python
chi_now = SPS.ssp_fit(n_MC=20)
```

```python
SPS.output_coeffs_MC_to_screen()
SPS.output_to_screen(block_plot=False)
```

```python
SPS.output_gas_emission(filename=join(output_path, out_file_elines))
SPS.output_fits(filename=join(output_path, out_file_fit))
SPS.output_coeffs_MC(filename=join(output_path, out_file_coeffs))
SPS.output(filename=join(output_path, out_file), block_plot=False)
```
