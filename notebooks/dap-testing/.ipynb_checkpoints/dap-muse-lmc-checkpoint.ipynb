```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set("talk", style="ticks")

from astropy.io import fits

from tqdm import tqdm

import contextlib
import sys

import warnings
warnings.filterwarnings("ignore")

class DummyFile(object):
    def write(self, x): pass

@contextlib.contextmanager
def nostdout():
    save_stdout = sys.stdout
    sys.stdout = DummyFile()
    yield
    sys.stdout = save_stdout
```

```python
inp_path = "../../_fitting-data/LMC-MUSE/"
obs_path = '_data/lmc/obs/'
out_path = '_data/lmc/out/'
fig_path = "_data/lmc/figs"
os.makedirs(obs_path, exist_ok=True)
os.makedirs(out_path, exist_ok=True)
os.makedirs(fig_path, exist_ok=True)


ori_rss = sorted([os.path.join(root,file) for root, subs, files in os.walk(inp_path) for file in files if file.startswith("CS.LMC_043") and file.endswith(".RSS.fits.gz")])
err_rss = [os.path.join(os.path.dirname(file), f"e_{os.path.basename(file)}") for file in ori_rss]
```

```python
from specutils import Spectrum1D
from specutils.manipulation import gaussian_smooth
import astropy.units as u

from pyFIT3D.common.tools import get_wave_from_header


new_muse_fwhm = 6.0
res_correction = np.sqrt((new_muse_fwhm)**2 - (2.2)**2)/2.355

no_error_file = False
seds_name = []
for ipoint in tqdm(range(len(ori_rss)), desc="writing spectra", ascii=True, unit="SED"):
    f = fits.open(ori_rss[ipoint], memmap=False)
    if not os.path.isfile(err_rss[ipoint]):
        no_error_file = True
    else:
        e = fits.open(err_rss[ipoint], memmap=False)

    wl = get_wave_from_header(f[0].header)
    dummy_err = np.abs(1e-7*np.ones_like(wl))
    
    for ised, (sed, err) in enumerate(zip(f[0].data, e[0].data)):
        out_file = "{}_{:03d}-sed.txt".format(os.path.basename(ori_rss[ipoint]).split(".")[1], ised)
        seds_name.append(out_file)
        sed_path = os.path.join(obs_path,out_file)
        
#         if os.path.isfile(sed_path): continue
        
        sed_cor = gaussian_smooth(Spectrum1D(spectral_axis=wl * u.AA, flux=sed * u.erg/u.s/u.AA), stddev=res_correction/np.diff(wl)[0]).flux.value
        sed_err = dummy_err# if no_error_file else np.where(err>0, err, dummy_err)

        np.savetxt(sed_path, np.column_stack((
            np.arange(wl.size, dtype=int),
            wl,
            sed_cor,
            sed_err
        )))
    no_error_file = False
```

```python
# using the initial values:
#
# Instrumental dispersion: 2.5 AA
sigma_inst = np.sqrt((new_muse_fwhm)**2-(2.5)**2)/2.355

# Velocity dispersion of the gas in AA
sigma_gas = 3.7

# Masks:
#     File with list of ranges of wavelength to be masked in whole analysis
mask_list = None#'../../_fitting-data/_configs/mask_elines.txt'

######################
######################
# Non linear analysis: (redshift, observed dispersion and dust attenuation)
#     wavelength range: 
#         [3850, 4700] AA for the redshift and sigma analysis
#         [3800, 7000] AA for the dust attenuation
w_min_max = [4800, 8000]
nl_w_min_max = [4800, 6000]

#     models:
#         Miles solar 3 models
ssp_nl_fit_file = '../../_fitting-data/_basis_mastar_v2/stellar-basis-spectra-5.fits.gz'

#     emission lines to be masked:
elines_mask_file = '../../_fitting-data/_configs/MaNGA/emission_lines_long_list.txt'

#     redshift:
#         initial value: 0.02449
#                 delta: 0.0001
#                   min: 0.02349
#                   max: 0.02549
redshift_set = [0.000875, 0, -0.5, 0.5]

#     dispersion: (km/s)
#         initial value: 30
#                 delta: 20
#                   min: 1
#                   max: 150
sigma_set = [0, 0, 0, 350]

#     dust attenuation at V band (mag):
#         initial value: 0.3
#                 delta: 0.15
#                   min: 0
#                   max: 1.6
AV_set = [0, 0.01, 0, 1.6]

ssp_file = '../../_fitting-data/_basis_mastar_v2/stellar-basis-spectra-100.fits.gz'
par_file = '../../_fitting-data/_basis_mastar_v2/stellar-basis-params-100.fits.gz'

#     configuration file:
config_file = '../../_fitting-data/_configs/auto_ssp_V500_several_Hb.config'

# plot options
stellar_param = fits.open(par_file, memmap=False)

labels = {"TEFF":r"$\log{T_\text{eff}}$", "LOGG":r"$\log{g}$", "MET":r"$[\text{Fe}/\text{H}]$", "ALPHAM":r"$[\alpha/\text{Fe}]$"}
cmap = "binary_r"
color = sns.color_palette(cmap)[0]
```

```python
import subprocess


subprocess.run(["tsp", "-S", "30"])
for sed_file in tqdm(seds_name, desc="fitting LMC", ascii=True, unit="SED"):
    label = sed_file.replace("-sed.txt", "")
#     if label != "LMC_043_001": continue
#     if os.path.isfile(os.path.join(out_path,label)): continue
    subprocess.run(f"tsp lvm-dap {os.path.join(obs_path,sed_file)} {ssp_file} {sigma_inst} {label} --rsp-nl-file {ssp_nl_fit_file} --w-range {' '.join(map(str,w_min_max))} --w-range-nl {' '.join(map(str,nl_w_min_max))} --redshift {' '.join(map(str,redshift_set))} --sigma {' '.join(map(str,sigma_set))} --AV {' '.join(map(str,AV_set))} --sigma-gas {sigma_gas} --emission-lines-file {elines_mask_file} -cd --output-path {out_path}".split(), capture_output=True)
```

```python
from lvmdap.analysis.plotting import plot_dap_fit


for sed_file in tqdm(seds_name, desc="plotting results", ascii=True, unit="SED"):
    label = sed_file.replace("-sed.txt", "")
    if label != "LMC_043_001": continue
    fig_file = os.path.join(fig_path, f"{label}.png")
#     if os.path.isfile(fig_file): continue
    
    model = pd.read_csv(os.path.join(out_path,f"coeffs_{label}"), sep="\s+")
    model.columns = model.columns[1:].tolist()+["#"]
    model.drop(columns="#", inplace=True)

    f = fits.open(os.path.join(out_path,f"output.{label}.fits.gz"), memmap=False)

    fig, ax0, axs = plot_dap_fit(
        f[0],
        model.COEFF.values,
        stellar_param,
        labels, cmap, color
    )
    fig.savefig(fig_file, bbox_inches="tight")
    plt.close(fig)
    f.close()
    del f
```
