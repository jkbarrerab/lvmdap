```python
import itertools
from copy import deepcopy as copy
from scipy.ndimage import median_filter
from pyFIT3D.common.tools import rss_seg2cube, get_slice, smooth_spec_clip_cube, spec_extract_cube_mean
from pyFIT3D.common.io import print_done, print_block_init


def mask_stars_in_FoV(name, V__yx, msk__yx, x0=None, y0=None, mask_val=0):
    time_ini_loc = print_block_init('Mask stars in FoV...')
    ny, nx = V__yx.shape
    ix0 = int(nx*.5) if x0 is None else x0
    iy0 = int(ny*.5) if y0 is None else y0
    V_msk__yx = np.ones_like(msk__yx)
    V_msk__yx[msk__yx == 2] = 0
    # look for stars in the field
    ix0, iy0 = int(nx*.5), int(ny*.5)
    val0 = V__yx[iy0, ix0]
    max_dist = 0.15*np.sqrt(nx**2 + ny**2)
    for ixy in itertools.product(range(nx),range(ny)):
        ix, iy = ixy
        val = V__yx[iy, ix]
        dist = np.sqrt((ix - ix0)**2 + (iy - iy0)**2)
        if (val > val0) & (dist > max_dist):
            nx0 = ix - 3
            nx1 = ix + 3
            ny0 = iy - 3
            ny1 = iy + 3
            # img border control
            nx0 = 0 if nx0 < 0 else nx0
            ny0 = 0 if ny0 < 0 else ny0
            nx1 = nx - 1 if nx1 > (nx - 1) else nx1
            ny1 = ny - 1 if ny1 > (ny - 1) else ny1
            # mask region
            for iixy in itertools.product(range(nx0, nx1),range(ny0, ny1)):
                iix, iiy = iixy
                V_msk__yx[iiy, iix] = mask_val
    # save mask to file
    print_done(time_ini=time_ini_loc)
    return V_msk__yx

# !echo "V 4800 5500" > slice_V.conf
# !cat slice_V.conf
# slice_conf = "/disk-a/sanchez/ppak/DATA/SSP/legacy/slice_V.conf"
slice_conf = "_data/lmc/configs/slice_V.conf"

slice_prefix = f'img_LMC_{POINTING}'
slices = get_slice(copy(org_cube__wyx), org_wave__w, slice_prefix, slice_conf)
V_img_slice_key = f'{slice_prefix}_V_4800_5500'
V__yx = copy(slices[V_img_slice_key])
V__yx[np.isnan(V__yx)] = -1

filt_V__yx = median_filter(V__yx, size=(2, 2), mode='reflect')
msk__yx = np.zeros_like(V__yx)
V_msk__yx = mask_stars_in_FoV(slice_prefix, copy(V__yx), msk__yx, mask_val=1)
mV__yx = filt_V__yx*V_msk__yx

org_rss__sw, _, _x_mean, _y_mean, npt_mean = spec_extract_cube_mean(copy(org_cube__wyx), seg_map__yx)
cube__wyx = rss_seg2cube(org_rss__sw, seg_map__yx)

slice_prefix = f'SEG_img_LMC_{POINTING}'
slices = get_slice(copy(cube__wyx), wave_rss__w, slice_prefix, slice_conf)
k = list(slices.keys())[0]
V_slice__yx = slices[k]
V_slice__yx[np.isnan(V_slice__yx)] = -1
scale_seg__yx = np.nan_to_num(mV__yx/V_slice__yx)


ssp_mod_tmp__wyx = rss_seg2cube(copy(output_rss__tsw[1]), seg_map__yx)
ssp_mod_cube__wyx = ssp_mod_tmp__wyx*scale_seg__yx
tmp_cube__wyx = org_cube__wyx - ssp_mod_cube__wyx

# smooth
smooth_cube__wyx = smooth_spec_clip_cube(copy(tmp_cube__wyx), wavebox_width=75, sigma=1.5, wavepix_min=10, wavepix_max=1860)

# generate GAS rss input
gas_cube__wyx = tmp_cube__wyx - smooth_cube__wyx
```

```python
from pyFIT3D.common.stats import pdl_stats, _STATS_POS
from pyFIT3D.common.tools import trim_waves
from pyFIT3D.common.auto_ssp_tools import sel_waves, ConfigAutoSSP, ConfigEmissionModel
from pyFIT3D.common.gas_tools import _EL_MODELS, fit_elines_main


def parallel_elines(ix, iy, wave__w, gas_cube__wyx, err_cube__wyx, configs_map__yx, elines_r):
    flux_elines_dict = {(iy,ix): dict.fromkeys(elines_r.keys(), 0.0)}
    
    if not configs_map__yx[iy, ix]: return flux_elines_dict

    flux__w = gas_cube__wyx[:, iy, ix]
    eflux__w = err_cube__wyx[:, iy, ix]

    st_eflux = pdl_stats(eflux__w)
    eflux__w = 0.5*(eflux__w + st_eflux[_STATS_POS['median']])

    med_SNR = np.nanmedian((flux__w/eflux__w))
    eflux__w = np.where(~np.isnan(eflux__w), eflux__w, (flux__w/med_SNR if med_SNR!=0 else 1e-3))

    auto_ssp_config = ConfigAutoSSP(configs_map__yx[iy, ix])
    for isys, cf in enumerate(auto_ssp_config.systems_config):
        # trim & apply masks
        sel_wl_range = trim_waves(wave__w, [auto_ssp_config.systems[isys]["start_w"], auto_ssp_config.systems[isys]["end_w"]])
        wave_msk__w = wave__w[sel_wl_range]
        flux_msk__w = flux__w[sel_wl_range]
        eflux_msk__w = eflux__w[sel_wl_range]

        with nostdout():
            EL = fit_elines_main(wavelength=wave_msk__w, flux=flux_msk__w, sigma_flux=eflux_msk__w,
                                config=cf, n_MC=20, n_loops=3, scale_ini=0.15,
                                fine_search=None, redefine_max=False,
                                max_factor=1.2, redefine_min=False,
                                plot=0, run_mode="RND", check_stats=False)

        for imod, model_type in enumerate(EL.config.model_types):
            # BUG: when the spectrum is so bad that no emission line fitting is performed, central wavelengths are reported as zeros.
            #      It should be reported the original wavelengths as those are not free parameters of the model.
            model_wl = EL.final_fit_params_mean[imod][_EL_MODELS[model_type]["central_wavelength"]]
            if model_type != "eline" or model_wl==0: continue
            flux_elines_dict[(iy,ix)][model_wl] = EL.final_fit_params_mean[imod][_EL_MODELS[model_type]["flux"]]
    
    return flux_elines_dict
```

```python
from scipy.interpolate import griddata
from joblib import Parallel, delayed


rss_config_files = sorted([os.path.join(out_path,file) for file in os.listdir(out_path) if file.startswith(f"LMC_{POINTING}") and file.endswith(".config")])

seg = copy(seg_map__yx)
yo, xo = np.where(seg>0)
values = seg[yo,xo]

yi, xi = np.where(seg<9999)

seg_int = griddata(np.column_stack((xo,yo)), values, np.column_stack((xi,yi)), method="nearest").reshape(seg.shape)*(~np.isnan(cube[0].data).all(axis=0))

# plt.figure(figsize=(10,10))
# plt.imshow(seg, origin="lower")
# plt.figure(figsize=(10,10))
# plt.imshow(seg_int, origin="lower")

configs_map__yx = np.asarray(rss_config_files, dtype=str)[seg_int.astype(int)-1]
configs_map__yx[seg_int==0] = ""

ny, nx = gas_cube__wyx.shape[1:]

parallel = Parallel(n_jobs=30, verbose=1)

elines_fit = parallel(delayed(parallel_elines)(ix, iy, org_wave__w, gas_cube__wyx, cube[1].data, configs_map__yx, elines_r) for ix, iy in itertools.product(range(nx),range(ny)))
```
