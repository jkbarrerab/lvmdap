```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm


sns.set("talk", style="ticks")

from astropy.io import fits

import contextlib
import sys

class DummyFile(object):
    def write(self, x): pass

@contextlib.contextmanager
def nostdout():
    save_stdout = sys.stdout
    sys.stdout = DummyFile()
    yield
    sys.stdout = save_stdout
```

```python
simulation_path = "../../_fitting-data/simulations/OrionBar_v0_1.fits"
f = fits.open(simulation_path)
```

```python
wl = ((np.arange(f[0].header["NAXIS1"])+1) - f[0].header["CRPIX1"]) * f[0].header["CDELT1"] + f[0].header["CRVAL1"]
```

```python
plt.figure(figsize=(15,5))
plt.semilogy(wl, f[0].data[-1])
```

```python
# no noise
# no LOSVD
# how to deal with kurtosis
# clowns method (works great for CO, github)
```

```python
from astropy.convolution import Gaussian1DKernel
from specutils import Spectrum1D
from specutils.manipulation import convolution_smooth
import astropy.units as u


BASENAME = os.path.basename(simulation_path).split(".")[0]
SIM_SINGLE_PATH = "_data/eric_sim"
if not os.path.isdir(SIM_SINGLE_PATH):
    os.mkdir(SIM_SINGLE_PATH)
    os.mkdir(os.path.join(SIM_SINGLE_PATH,"obs"))
    os.mkdir(os.path.join(SIM_SINGLE_PATH,"out"))

new_sim_res = 4.0
gaussian_kernel = Gaussian1DKernel(stddev=new_sim_res/2.355/0.48)

obs_list = []
for i, sed in tqdm(enumerate(f[0].data), total=len(f[0].data), desc="adding instrumental effects", ascii=True):

    sed_inst = convolution_smooth(Spectrum1D(spectral_axis=wl * u.AA, flux=f[0].data[0] * u.erg/u.s/u.AA), gaussian_kernel)
    sed_inst = convolution_smooth(sed_inst, Gaussian1DKernel(stddev=1.5/0.48)).flux.value
    
    sed_error = np.ones_like(sed_inst)*sed_inst[np.argmin(np.abs(6000-wl))]/20
    sed_noisy = sed_inst + np.random.randn(sed_inst.size)*sed_error
    
    sim_obs_path = os.path.join(os.path.join(SIM_SINGLE_PATH,"obs"), f"{BASENAME}-{i}.txt")
#     np.savetxt(sim_obs_path, np.column_stack((np.arange(wl.size), wl, sed_noisy, sed_error)))
    
    obs_list.append(sim_obs_path)
```

```python
from pyFIT3D.common.io import create_ConfigAutoSSP_from_lists
from pyFIT3D.common.io import create_emission_lines_file_from_list
from pyFIT3D.common.gas_tools import detect_create_ConfigEmissionModel
from pyFIT3D.common.io import create_emission_lines_mask_file_from_list


configs_path = os.path.join(SIM_SINGLE_PATH, "configs")
redshift = 0
sigma_guess = new_sim_res/2.355
chunks = [[3600, 5500], [5500, 7000], [7000, 9800]]
peak_find_nsearch = 1
peak_find_threshold = 0.5
peak_find_dmin = 1
plot = 0
crossmatch_list_filename = None
crossmatch_list_filename = "../../_fitting-data/_configs/MaNGA/emission_lines_long_list.MaNGA"
#crossmatch_list_filename = 'emission_lines.txt'
crossmatch_min_dist_AA = 6
sort_by_flux = True
verbose = 2

r = detect_create_ConfigEmissionModel(
    
    # spectrum
    wl, sed_noisy, 
    
    redshift=redshift,
    sigma_guess=sigma_guess,
    
    # Chunks to be analyzed (i.e. the chunks of the generated ConfigEmissionModel files)
    chunks=chunks,
    
    # Add a polynomial model to the generated configs
    polynomial_order=1,
    polynomial_coeff_guess=[0.000, 0.001],
    polynomial_coeff_boundaries=[[-1e13, 1e13], [-1e13, 1e13]],

#     polynomial_order=1,
#     polynomial_coeff_guess=[0.001],
#     polynomial_coeff_boundaries=[-1e13, 1e13],

    ########################################################
    ### DEFAULT VALUES, NOT NEED TO BE USED FOR THOSE VALUES
    flux_boundaries_fact=[0.001, 1000],
    sigma_boundaries_fact=[0.1, 1.5],
    v0_boundaries_add=[-50, 50],
    ########################################################
    
    # Peak find options
    peak_find_nsearch=peak_find_nsearch,
    peak_find_threshold=peak_find_threshold,
    peak_find_dmin=peak_find_dmin,
    
    
    # Crossmatch options
    crossmatch_list_filename=crossmatch_list_filename,
    crossmatch_absdmax_AA=crossmatch_min_dist_AA,

    # Nice plots
    plot=plot,
    
    output_path=configs_path,
    sort_by_flux=sort_by_flux,
    
    verbose=verbose,
)
config_filenames, wl_chunks, wave_peaks_tot, wave_peaks_tot_corr = r

create_emission_lines_mask_file_from_list(wave_peaks_tot, eline_half_range=8*sigma_guess, output_path=configs_path)
create_emission_lines_file_from_list(wave_peaks_tot, output_path=configs_path)
create_ConfigAutoSSP_from_lists(wl_chunks, config_filenames, output_path=configs_path)
```

```python
OUTPUT_PATH = os.path.join(SIM_SINGLE_PATH,"out")

STELLAR_BASIS = "../../_fitting-data/_basis_mastar_v2/stellar-basis-spectra-100.fits.gz"
STELLAR_PARAMS = "../../_fitting-data/_basis_mastar_v2/stellar-basis-params-100.fits.gz"
SIGMA_INST = np.sqrt((new_sim_res/2.355)**2 - (2.62/2.355)**2)
MASK_ELINES = os.path.join(configs_path, "autodetect.mask_elines.txt")
CONFIG = os.path.join(configs_path, "autodetect.auto_ssp_several.config")
EMISSION_LINES = "../../_fitting-data/_configs/MaNGA/emission_lines_long_list.MaNGA"
```

```python
from lvmdap._cmdline import dap
from lvmdap.analysis.plotting import plot_triang_pdfs


# fit the Eric's spectra
for target_sed in tqdm(obs_list[-1:], desc="fitting spectra", ascii=True, unit="SED"):
    label = os.path.basename(target_sed).split(".")[0]
    
    cmdline = f"{target_sed} {STELLAR_BASIS} {SIGMA_INST} {label} {MASK_ELINES} {CONFIG} --emission-lines {EMISSION_LINES} --w-range 3600 9799 --w-range-nl 3600 4700 --redshift 0 0 0 0 --sigma 0 0 0 0 --AV 0 1 0 10 -o {OUTPUT_PATH} -c".split()
    with nostdout():
        dap._main(cmdline)
```

```python
LABEL = "OrionBar_v0_1-423"

models = fits.open(f"{OUTPUT_PATH}/output.{LABEL}.fits.gz")
models[0].header

wl = np.array([models[0].header["CRVAL1"] + (i*models[0].header["CDELT1"]) for i in range(models[0].header["NAXIS1"])])
fl_o = models[0].data[0]
fl_m = models[0].data[1]
fl_j = models[0].data[2]
fl_g = models[0].data[3]
fl_r = models[0].data[4]
fl_n = models[0].data[5]

elines = {4101.0: r"$\text{H}\delta$", 4340.468: r"$\text{H}\gamma$", 4861.32: r"$\text{H}\beta$", 4958.91: r"$[\text{O}\textsc{iii}]\lambda4959$",
          5006.84: r"$[\text{O}\textsc{iii}]\lambda5007$", 6562.817: r"$\text{H}\alpha$", 6583.6: r"$[\text{N}\textsc{ii}]\lambda6584$",
          6716.47: r"$[\text{S}\textsc{ii}]\lambda6717$", 6730.85: r"$[\text{S}\textsc{ii}]\lambda6731$"}

fig, (ax1,ax2) = plt.subplots(2, figsize=(20,10), sharex=True, sharey=False)
xlims = plt.xlim(3600,9800)

# plt.xlim(4580,5100)
# plt.xlim(6400,6800)

mask_a1 = (xlims[0]<=wl)&(wl<=xlims[1])
ax1.plot(wl[mask_a1], fl_o[mask_a1], "-", color="0.7", lw=3)
ax1.plot(wl[mask_a1], fl_j[mask_a1], "-", color="tab:purple")
# ax1.set_ylim(0,0.2)
ax1.set_ylabel(r"$F_\lambda$")
_, ymax = ax1.get_ylim()
for ewl in elines:
    ax1.axvline(ewl, ls="--", lw=1, color="0.2")
    ax2.axvline(ewl, ls="--", lw=1, color="0.2", zorder=999)
    if ewl in [5006.84, 6583.6, 6730.85]:
        ax1.text(ewl+17, ymax-ymax*0.05, elines[ewl], ha="left", va="top", fontsize="small", rotation=90, color="0.2")
    else:
        ax1.text(ewl-8, ymax-ymax*0.05, elines[ewl], ha="right", va="top", fontsize="small", rotation=90, color="0.2")

axins1 = ax1.inset_axes([0.20, 0.25, 0.1, 0.6])
axins1.set_xlim(4840,4880)
mask_ai1 = (4840<=wl)&(wl<=4880)
axins1.plot(wl[mask_ai1], fl_o[mask_ai1], "-", color="0.7", lw=3)
axins1.plot(wl[mask_ai1], fl_j[mask_ai1], "-", color="tab:purple")
# axins1.set_ylim(0,5)
axins1.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
# ax1.indicate_inset_zoom(axins1)

axins2 = ax1.inset_axes([0.65, 0.25, 0.1, 0.6])
# axins2.set_xlim(6700,6750)
# mask_ai2 = (6700<=wl)&(wl<=6750)
axins2.set_xlim(6540,6595)
mask_ai2 = (6540<=wl)&(wl<=6595)
axins2.plot(wl[mask_ai2], fl_o[mask_ai2], "-", color="0.7", lw=3)
axins2.plot(wl[mask_ai2], fl_j[mask_ai2], "-", color="tab:purple")
# axins2.set_ylim(0,1.5)
axins2.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
# ax1.indicate_inset_zoom(axins2)

ax2.plot(wl[mask_a1], (fl_o-fl_g)[mask_a1], "-", color="tab:blue", zorder=1)
ax2_ylims = ax2.get_ylim()
ax2.plot(wl[mask_a1], (fl_o)[mask_a1], "-", color="0.7", lw=3, zorder=-1)
ax2.set_ylim(ax2_ylims)
ax2.set_xlabel(r"$\lambda$")
ax2.set_ylabel(r"$F_\lambda$")

fig.tight_layout()
```
